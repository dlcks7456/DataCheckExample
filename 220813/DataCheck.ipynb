```python
# max
dc.df['Q3x2_max'] = dc.df[Q3x2].max(axis=1)
```

```python
cols = Q3x2 + ['Q3x2_max']
dc.df[cols]
```

```python
# HQ3 logic check
idxs = list(dc.df.index)
for idx in idxs :
    max_value = dc.df.loc[idx, 'Q3x2_max']
    HQ3_value = dc.df.loc[idx, 'HQ3']
    base_qid = f'Q3x2b{HQ3_value}'
    base_value = dc.df.loc[idx, base_qid]
    if not max_value == base_value :
        print(list(dc.df.loc[idx, ['record', base_qid, 'Q3x2_max']]))
```
